# Generate Summary Automatically

```python
duplicate_amfi = fund_master['amfi_code'].duplicated().sum()

missing_nav = nav_history['nav'].isnull().sum()

duplicate_rows = nav_history.duplicated(
    subset=['amfi_code','date']
).sum()

master_codes = set(fund_master['amfi_code'])
nav_codes = set(nav_history['amfi_code'])

missing_codes = master_codes - nav_codes

found_pct = (
    len(master_codes & nav_codes)
    / len(master_codes)
) * 100

print("\nDATA QUALITY SUMMARY")

print("\n1. fund_master.csv")
print(f"   - Duplicate AMFI Codes: {duplicate_amfi}")

if (
    fund_master['category'].isnull().sum() == 0
    and fund_master['sub_category'].isnull().sum() == 0
    and fund_master['risk_category'].isnull().sum() == 0
):
    print("   - Category information complete")

print("\n2. nav_history.csv")
print(f"   - Missing NAV values: {missing_nav}")
print(f"   - Duplicate rows: {duplicate_rows}")

print("\n3. AMFI Validation")
print(f"   - {found_pct:.2f}% AMFI codes found")
print(f"   - {len(missing_codes)} codes missing NAV history")
```
